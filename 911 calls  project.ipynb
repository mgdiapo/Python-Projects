```python
dfccfinalv2 = pd.merge(dfccfinal,dfcbr, 'right',on='Municipality')
dfccfinalv2.head()
```

```python
dfccfinalv2['EMS_Most_Common'] = np.where((dfccfinalv2['EMS'] > dfccfinalv2['Fire']) & (dfccfinalv2['EMS'] > dfccfinalv2['Traffic']), 
                                           'yes', 'no')
dfccfinalv2['Fire_Most_Common'] = np.where((dfccfinalv2['Fire'] > dfccfinalv2['EMS']) & (dfccfinalv2['Fire'] > dfccfinalv2['Traffic']), 
                                           'yes', 'no')
dfccfinalv2['Traffic_Most_Common'] = np.where((dfccfinalv2['Traffic'] > dfccfinalv2['Fire']) & (dfccfinalv2['Traffic'] > dfccfinalv2['EMS']), 
                                           'yes', 'no')
dfccfinalv2['EMS_Likely'] = np.where((dfccfinalv2['EMS'] > np.mean(dfccfinalv2['EMS'])),  
                                           'yes', 'no')
dfccfinalv2['Fire_Likely'] = np.where((dfccfinalv2['Fire'] > np.mean(dfccfinalv2['Fire'])),  
                                           'yes', 'no')
dfccfinalv2['Traffic_Likely'] = np.where((dfccfinalv2['Traffic'] > np.mean(dfccfinalv2['Traffic'])),  
                                           'yes', 'no')
```
